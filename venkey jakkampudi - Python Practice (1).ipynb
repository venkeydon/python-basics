# Type conversions
There are two types of type conversions

```python
# Implicit type conversions
a=10
b=5.4
c=a+b
print(' a is of type:',type(a))
print('b is of type:',type(b))
print(c)
print('c is of type:',type(c))
```

```python
# Explicit type conversions
#converting to int
print(int(3.14))   #from float to int
a=int('101')
b=int('202')      #from numeric string to int
c=a+b
print(c)
print(int('1010',2)) #from binary to decimal int
print(int('341',8))  #from octal to decimal int
print(oct(225))
print(int('21',16))   #from hexa to decimal int
print(hex(33))

#convert to float
print(float(3))       #from int to float
i=float('4.8')
j=float('5.3')        #from numerical string to float
k=i+j
print(k)

#conver to complex
print(complex(30))     #from int to complex
x=complex(4.85,1.1)
y=complex(7.68,2.1)    #from numeric string to complex
z=x+y
print(z)

#covert to bool
print(bool(35))       #from int to bool
print(bool(3.14))     #from float to  bool
```

##If-else Statements

-->if condition:
      statement(s)
-->else:
      statement(s)

```python
#using if else statements
qty=int(input('enter value of quantiy:'))         #input the number of quantity
price=float(input('enter value of price:'))       #input the price
if qty>=1000:                                     #if condition
    dis=10                                        #statement
else:                                             #else
    dis=0                                         # statement
totexp=(qty*price)-(qty*price*dis/100)            #here we find the total expenses by subing the discount
print('Total expenses=Rs.'+str(totexp))           #concatinating result with the print statement
```

```python
#creating a function
def is_leap(year):
 if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    print(year,"is a Leap Year")
 else:
    print(year," is not a leap year")
year = int(input())
print(is_leap(year))
```

```python
#recursive factorial function
def refact(n):
  if n==0:
    return 1
  else:
    p=n*refact(n-1)
    return p
num=int(input('enter any number'))
fact=refact(num)
print('factorial is:',fact)
```
